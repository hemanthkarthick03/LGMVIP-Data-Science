```python
data['Wound'] = data['Wound'].fillna(0)
data['kill'] = data['kill'].fillna(0)
data['Casualities'] = data['kill'] + data['Wound']
```

```python
#AttacksOnYears

year = data['Year'].unique()
years_count = data['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year,
           y = years_count,
           palette = "tab10")
plt.xticks(rotation = 60)
plt.xlabel('Year',fontsize=20)
plt.ylabel('No. of Attacks on Each Year',fontsize=20)
plt.title('Attacks vs Years',fontsize=30)
plt.show()
```

```python
#AttacksOnMonths

plt.figure(figsize=(20, 10))
sns.countplot('Month', data=df, palette = "tab10", edgecolor='black')
plt.title('Attacks vs Months',fontsize=20)
plt.xlabel('Month', fontsize=16)
plt.ylabel('No. of Attacks on Each Month', fontsize=16)
plt.show()
```

```python
#AttacksOnDays

plt.figure(figsize=(20, 10))
sns.countplot('day', data=df, palette = "tab10", edgecolor=sns.color_palette("YlOrBr", 10))
plt.title('Attacks vs Days', fontsize=20)
plt.xlabel('Day', fontsize=16)
plt.ylabel('No. of Attacks on Each Days', fontsize=16)
plt.show()
```

```python
pd.crosstab(data.Year, data.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities By Region vs Year',fontsize=25)
plt.ylabel('No. of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

```python
#PieChart
region_of_attack = df['Region'].value_counts()
year_of_attack = df['Year']

plt.figure(figsize=(5,5))
plt.pie(region_of_attack.values, labels=None, autopct='%.0f%%')
label_colours = sns.color_palette('bright', len(region_of_attack))
legend_labels = [f'{region} ({count})' for region, count in zip(region_of_attack.index, region_of_attack.values)] # zip makes tuples of length two given the iterables
plt.legend(labels=legend_labels, prop={'size': 8}, bbox_to_anchor=(0.85, 0.15), facecolor='white')
plt.title('Terrorist Attacks by Region 1970-2017')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(df['Group'].value_counts()[1:16].values, df['Group'].value_counts()[1:16].index, palette = "tab10", orient='h')
plt.title('No. of Terrorist Activities by Terrorist Group', fontsize=20)
plt.xlabel('Groups', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.show()
```

```python
region_attacks = df.groupby(['Region', 'Year']).size().reset_index(name='counts')
for region in region_attacks['Region'].unique():
    region_data = region_attacks[region_attacks['Region'] == region] 
    plt.plot(region_data['Year'], region_data['counts'], label=region)

plt.title('Number of Attacks by Region and Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(data["Targettype"],order=data['Targettype'].value_counts().index, palette = "tab10", edgecolor=sns.color_palette("YlOrBr", 10));
plt.xticks(rotation=90)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Attack per year",fontsize=20)
plt.show()
```

```python
data[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='bar',color='brown', figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("No of killed people",fontsize=15)
plt.show()
```
